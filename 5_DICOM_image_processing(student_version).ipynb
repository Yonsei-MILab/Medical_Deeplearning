#**5-2.1 Importing Functions**



```
import pydicom
import numpy as np
import sys, os
import difflib
import matplotlib.pyplot as plt
from scipy import ndimage
```

#**5-2.2 Reading a DICOM file (ex. Sec_100.mag)**


```
tmp_DICOM = pydicom.read_file('Sec_100.mag')
tmp_DICOM
```

```
print(tmp_DICOM.SliceLocation) 
```

#**5-2.3. Reading DICOM image**

```
img = tmp_DICOM.pixel_array
print(img.shape) # Shape
print(img.dtype) # type
```

#**5-2.4. Copying DICOM image to numpy array**

```
img = np.array(img, dtype='float32').copy()
print(img.shape)
print(img.dtype)
print(img.max())
print(img.min())

plt.figure(figsize=(10,10))
plt.imshow(img,cmap='gray')
plt.show()
```
